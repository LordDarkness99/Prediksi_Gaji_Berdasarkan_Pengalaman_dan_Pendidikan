```python
import numpy as np
import matplotlib.pyplot as plt
```

DATASET

```python
X = np.array([
    [1, 1],
    [2, 1],
    [2, 2],    [3, 2],
    [3, 3],
    [4, 3],
    [5, 3],
    [6, 4],
    [7, 4],
    [8, 4]
])

y = np.array([4000000, 5000000, 6000000, 7000000, 8000000, 9000000, 
              10000000, 12000000, 15000000, 17000000])

m, n = X.shape

print(f"Jumlah data: {m}, Jumlah fitur: {n}")
```

NORMALISASI FITUR

```python
X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)
X_norm = (X - X_mean) / X_std

# Tambahkan kolom 1 untuk bias (intercept)
X_train = np.hstack([np.ones((m, 1)), X_norm])
```

DEFINIISI FUNGSI MODEL LINIER DAN FUNGSI BIAYA

```python
def predict(X, w):
    return np.dot(X, w)
    
def compute_cost(X, y, w):
    m = len(y)
    cost = (1 / (2 * m)) * np.sum((predict(X, w) - y) ** 2)
    return cost
```

GRADIENT DESCENT

```python
def gradient_descent(X, y, w, alpha, iterations):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        predictions = predict(X, w)
        errors = predictions - y
        gradients = (1 / m) * np.dot(X.T, errors)
        w = w - alpha * gradients
        cost = compute_cost(X, y, w)
        cost_history.append(cost)

        if i % 100 == 0:
            print(f"Iterasi {i}: Cost = {cost:.2f}")
    return w, cost_history
```

```python
w_init = np.zeros(n + 1)
alpha = 0.1
iterations = 1000
```

TRAINING MODEL

```python
w_final, cost_history = gradient_descent(X_train, y, w_init, alpha, iterations)

print("\nâœ… Training Selesai!")
print(f"Parameter model (w): {w_final}")
print(f"Biaya akhir: {cost_history[-1]:.2f}")

# ================================================================
# ðŸ“ˆ 9. Visualisasi Penurunan Biaya
plt.figure(figsize=(8, 4))
plt.plot(range(iterations), cost_history, 'r')
plt.xlabel("Iterasi")
plt.ylabel("Nilai Cost (J)")
plt.title("Penurunan Nilai Cost selama Training")
plt.grid(True)
plt.show()
```

PREDIKSI GAJI

```python
def prediksi_gaji(pengalaman, pendidikan):
    x_input = np.array([[1, pengalaman, pendidikan]])
    x_input[:, 1:] = (x_input[:, 1:] - X_mean) / X_std
    return predict(x_input, w_final)[0]

# Contoh prediksi
tahun = 5
pendidikan = 3  # S1
gaji_prediksi = prediksi_gaji(tahun, pendidikan)

print(f"\nðŸ“Š Prediksi Gaji untuk {tahun} tahun pengalaman dan pendidikan S1:")
print(f"â‰ˆ Rp {gaji_prediksi:,.0f}")
```
